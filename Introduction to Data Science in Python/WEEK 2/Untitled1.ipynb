```python
df=pd.read_csv('datasets/presidents.csv',index_col=0)
```

```python
df.head(10)
```

```python
df['First']=df['President']
df['First']=df['First'].replace('[ ]{1}.*','',regex=True)
```

```python
df.head(10)
```

```python
del(df['First'])

def splitname(row):
    row['First']=row['President'].split(' ')[0]
    row['Last']=row['President'].split(' ')[-1]
    return row
```

```python
df=df.apply(splitname,axis="columns")
```

```python
df.head(10)
```

```python
del(df['First'])
del(df['Last'])

pattern="(^[\w]*)(?:.* )([\w]*$)"
df['President'].str.extract(pattern).head()
#([\w]*$)
```

```python
pattern="(?P<first>^[\w]*)(?:.* )(?P<last>[\w]*$)"
name=df['President'].str.extract(pattern)
#([\w]*$)
name
```

```python
df['First']=name['first']
df['Last']=name['last']
df.head(10)
```
